```python
import numpy as np                               
import pandas as pd                             
```

```python
balasore_fy_data_df = pd.read_csv('data/bls_fy_2014_2018.csv', parse_dates=['allotment_date'])
department_mapping_df = pd.read_csv('data/list_of_departments.csv')
```

```python
balasore_fy_data_df
```

```python
balasore_fy_data_df["department_code"] = balasore_fy_data_df["ddo_code"].str[3:6]
```

```python
balasore_fy_data_df
```

```python
department_mapping_df
```

```python
balasore_fy_data_df = pd.merge(balasore_fy_data_df, department_mapping_df, on="department_code", how='outer')
```

```python
balasore_fy_data_df
```

```python
balasore_fy_data_df["major_head"] = balasore_fy_data_df["budget_code"].str.split("-", expand=True)[1]
balasore_fy_data_df["sub_major_head"] = balasore_fy_data_df["budget_code"].str.split("-", expand=True)[2]
balasore_fy_data_df["minor_head"] = balasore_fy_data_df["budget_code"].str.split("-", expand=True)[3]
balasore_fy_data_df["sub_minor_head"] = balasore_fy_data_df["budget_code"].str.split("-", expand=True)[4]
balasore_fy_data_df["detailed_head"] = balasore_fy_data_df["budget_code"].str.split("-", expand=True)[5]
balasore_fy_data_df["object_head"] = balasore_fy_data_df["budget_code"].str.split("-", expand=True)[6]
```

```python
balasore_fy_data_df
```

```python
balasore_fy_data_df.to_csv("data/balasore_fy_2014_2018_cleaned.csv", encoding='utf-8', index=False)
```
